```python
from bs4 import BeautifulSoup
import undetected_chromedriver as uc
import time
import urllib.parse

def get_memes_by_keyword(keyword, limit=10):
    options = uc.ChromeOptions()
    options.headless = True
    driver = uc.Chrome(options=options)

    encoded_keyword = urllib.parse.quote(keyword)
    search_url = f"https://memes.tw/search?q={encoded_keyword}"
    driver.get(search_url)
    time.sleep(3)

    for i in range(3):
        print(f"🔄 第 {i+1} 次滾動")
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    meme_elements = soup.select('img')  # 根據實際搜尋結果頁結構調整 selector
    print(f"🖼️ 抓到 {len(meme_elements)} 張圖片")

    memes = []
    for img in meme_elements:
        url = img.get('data-src') or img.get('src')
        if url and url.startswith('//'):
            url = 'https:' + url
        if url:
            # 可加入更多過濾條件
            if "memes.tw" in url:
                print(f"發現圖片：{url}")
                memes.append(url)
        if len(memes) >= limit:
            break

    driver.quit()
    return memes

if __name__ == "__main__":
    keyword = input("請輸入搜尋關鍵字：")
    urls = get_memes_by_keyword(keyword, limit=10)
    print("\n🔥 搜尋結果梗圖圖片連結：")
    for idx, url in enumerate(urls, 1):
        print(f"{idx}: {url}")
```

```python
from bs4 import BeautifulSoup
import undetected_chromedriver as uc
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def get_memes_tw_images(limit=10):
    options = uc.ChromeOptions()
    options.headless = True
    driver = uc.Chrome(options=options)

    driver.get("https://memes.tw/wtf")

    try:
        # 等待第一張圖片出現
        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "img"))
        )

        # 多次滾動觸發懶載入
        for _ in range(3):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(3)

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        meme_elements = soup.select('img')
        print(f"找到圖片數量：{len(meme_elements)}")

        memes = []
        for img in meme_elements:
            url = img.get('data-src') or img.get('src')
            if url and url.startswith('//'):
                url = 'https:' + url
            if url and "memes.tw" in url:
                memes.append(url)
            if len(memes) >= limit:
                break

    except Exception as e:
        print("錯誤:", e)
        memes = []

    driver.quit()
    return memes

if __name__ == "__main__":
    urls = get_memes_tw_images(limit=10)
    print("🔥 熱門梗圖圖片連結：")
    for idx, url in enumerate(urls, 1):
        print(f"{idx}: {url}")
```
